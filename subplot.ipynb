```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
score=np.array([78,61,82,53])
student=np.array(["I","me","myself","my"])

plt.subplot(2,1,1)
plt.plot(student,score)
plt.title("Line Chart")

#plot2
plt.title("Bar Graph")
plt.subplot(2,1,2)
plt.bar(student,score) 

plt.suptitle("student data")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

N = 100
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)

plt.scatter(x, y, c='green')
plt.plot(np.linspace(0, 1, 10), np.power(np.linspace(0, 1, 10), 2), c= "red", marker='.', linestyle=':')

plt.gca().invert_yaxis()
plt.show()
```
